```python
class ListNode:
  def __init__(self, list1, node, g, h):
    self.node_list = []
    for element in list1:
      self.node_list.append(element)
    self.node_list.append(node)
    
    self.g = g
    self.f = int(g) + int(h)
    self.ID = node
    
  def is_Goal(self, complete_nodes) :
    if complete_nodes in self.node_list:
      return True
    return False
  
def return_minimum_value_node(openList):
  min = 100000
  min_node = openList[0]
    
  for node in openList:
    if node.f < min:
      min = node.f
      min_node = node
  return min_node
  
if __name__ == "__main__":
  n = int(input("Enter total number of nodes"))
  print('Nodes will be numbered as 1 to ',n)

  print('Enter graph')

  Graph = []

  total_nodes = []
  for i in range(1,n+1):
      total_nodes.append(i)
        
  """
    for i in range(1, n+1):
      
        current = []
      
        for j in range(1, n+1):
        
            val = input()
            current.append(val)
        
        graph.append(current)
        print("row appended")
  """
    
  Graph=[
     [ 0 ,1,-1 ,4,-1],
     [-1, 0, 5, 2, 12],
     [ -1, -1, 0 ,-1, 3],
     [-1,-1, 2, 0 ,-1],
     [-1,-1,-1,-1, 0]

  ]
    
  start = int(input('Enter the number of the start node : '))
    
  print('Enter heuristic value for every node : ')
    
  heuristic = []
  for i in range(1, n+1):
    heu = int(input())
    dict = {i:heu}
    heuristic.append(dict)
      
  open = []
    
  heu_val = heuristic[start-1]
  heu_val = heu_val[start]
  print('heu val of start = ',heu_val)
    
  empty = []
    
  start_node = ListNode(empty, start, 0, heu_val)
    
  open.append(start_node)
    
  while len(open) > 0:
    current_node = return_minimum_value_node(open)
      
    open.remove(current_node)
      
    print('Current min node with value of f as = ', current_node.f)
      
    if current_node.is_Goal(n):
      print('Current node is the goal node')
      print('Optimal value = ', current_node.f)
        
      for i in current_node.node_list:
        print(i, end=" ")
      break
        
    else:
      print('current node is not goal node')
      adjacent_nodes = Graph[current_node.ID-1]
      count = 0
      for i in adjacent_nodes:
          
        if i != 0 and i != -1:
          print('current node = ', current_node.ID, ' successor = ', count + 1)
          heu_val = heuristic[count]
          heu_val = heu_val[count + 1]
          successor_node = ListNode(current_node.node_list, count+1, current_node.g+i, heu_val)
          open.append(successor_node)
        count += 1
    
    
    
    
    
```
